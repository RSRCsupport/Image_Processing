### Subsetting by shapefile

A shapefile was previosly created (i.e. shape.shp located where the image is) to crop the image to include only land areas. The next code will use [gdalwarp](https://www.gdal.org/gdalwarp.html) to subset the image spatially 

EXPLORE THIS http://www.acgeospatial.co.uk/jupyter-notebooks-and-satellite-imagery/..

```python
def subset(shp, img):
    outimg='Dunwich_20140701_WV2_raw8cut'+'.tif'
    cmd="gdalwarp -of GTiff -cutline %s -crop_to_cutline -dstalpha %s  %s" % (shp, img, outimg)
    os.system(cmd)
  
```

```python
shape='shape.shp'
```

```python
subset(shape, filename)
```

Next we will use the tool [gdal_translate](http://www.gdal.org/gdal_translate.html) within a function in Python to subset the image to 4 bands (RGBN) and change the format (from.img to tif). This will create a new image named "LandsatETM_20010822_4B.ti"

```python
def convertimg(infile):
    cmd= "gdal_translate -b 1 -b 2 -b 3 -b 4 %s LandsatETM_20010822_4B.tif" %(infile)
    os.system(cmd)
```

```python
convertimg(filename)
```
